Misalkan diberikan prosedur iterasi:


xr+1 =(2xr+3)**(0,5)
, r = 0, 1, 2, 3, . . .
Iterasi dihentikan jika memenuhi:\
|εRA| < εs\
Diketahui:\
x0 = 3,2\
εs = 0.005\
Tentukan hasil iterasinya.

```python
# Metode Iterasi Tetap
# Menghitung galat relatif hampiran

# Data awal
x0 = 3.2
eps_s = 0.005

# Inisialisasi
xr = x0
error = 1
r = 0
print (" Iterasi | xr+1 | Error ")
print (" -----------------------------")

while error > eps_s :

# Rumus iterasi
    x_next = (2*xr+3)**(0.5)

# Galat relatif hampiran
    error = abs (( x_next - xr) / x_next )
# Tampilkan hasil
    print (f"{r:3d} | { x_next :.5f} | { error :.5f}")

 # Update
    xr = x_next
    r += 1

print ("Iterasi berhenti ")
print (" Hasil akhir x =", round (xr , 5))
```
